```python
# -*- coding: utf-8 -*-
"""
CNN_Pneumonia_Detection_Optimized_v2.ipynb

Kode lengkap untuk melatih model klasifikasi gambar rontgen dada
dengan perbaikan TypeError pada class_weight.
"""

# ==============================================================================
# 1. IMPORT LIBRARY YANG DIBUTUHKAN
# ==============================================================================
import os
import random
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import kagglehub
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout,
    Rescaling, BatchNormalization, Activation, RandomFlip,
    RandomRotation, RandomZoom, RandomContrast
)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
from tensorflow.keras.preprocessing import image


# ==============================================================================
# 2. SETUP AWAL: DOWNLOAD DAN PERSIAPAN DATASET
# ==============================================================================
print("Downloading dataset from Kaggle Hub...")
# Menggunakan 'silent' untuk mengurangi output yang tidak perlu
base_dir = kagglehub.dataset_download("paultimothymooney/chest-xray-pneumonia")
data_root = os.path.join(base_dir, 'chest_xray', 'chest_xray')

# Definisikan path untuk setiap set
train_dir = os.path.join(data_root, 'train')
val_dir = os.path.join(data_root, 'val')
test_dir = os.path.join(data_root, 'test')

print("Dataset Path:", data_root)
print("Isi folder train:", os.listdir(train_dir))


# ==============================================================================
# 3. MEMBUAT PIPELINE DATA (tf.data)
# ==============================================================================
batch_size = 32
img_size = (150, 150)

def create_dataset(directory, shuffle=False):
    """Fungsi untuk membuat tf.data.Dataset dari direktori gambar."""
    ds = tf.keras.utils.image_dataset_from_directory(
        directory,
        labels='inferred',
        label_mode='binary',
        image_size=img_size,
        batch_size=batch_size,
        shuffle=shuffle,
        # Pastikan urutan kelas konsisten
        class_names=['NORMAL', 'PNEUMONIA']
    )
    class_names = ds.class_names
    # Optimasi pipeline dengan cache dan prefetch untuk performa tinggi
    return ds.cache().prefetch(buffer_size=tf.data.AUTOTUNE), class_names

# Membuat generator untuk setiap set data
train_generator, class_names = create_dataset(train_dir, shuffle=True)
val_generator, _ = create_dataset(val_dir)
test_generator, _ = create_dataset(test_dir)

print("\nPipeline data berhasil dibuat!")
print("Kelas terdeteksi:", class_names)
print("Label mapping:", dict(zip(class_names, range(len(class_names)))))


# ==============================================================================
# 4. STRATEGI 1: MENANGANI CLASS IMBALANCE
# ==============================================================================
# Ekstrak semua label dari data training untuk menghitung bobot kelas
train_labels = np.concatenate([y for x, y in train_generator], axis=0)

# Hitung bobot menggunakan scikit-learn
weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_labels),
    y=train_labels.flatten() # Meratakan label
)
class_weights_float = dict(zip(np.unique(train_labels), weights))

### PERBAIKAN DI SINI: Ubah keys dari float ke integer ###
class_weights = {int(k): v for k, v in class_weights_float.items()}

print(f"\nClass Weights yang akan digunakan (kunci sudah integer): {class_weights}")
print("Model akan memberikan 'denda' lebih besar jika salah memprediksi kelas 'NORMAL'.")


# ==============================================================================
# 5. STRATEGI 2: MENYEMPURNAKAN ARSITEKTUR MODEL
# ==============================================================================
# Definisikan layer augmentasi untuk diletakkan di dalam model
data_augmentation_layers = Sequential([
    RandomFlip("horizontal"),
    RandomRotation(0.1),
    RandomZoom(0.2),
    RandomContrast(0.2),
], name='data_augmentation')

# Definisikan model dengan Batch Normalization
model = Sequential([
    # Input layer
    Input(shape=(img_size[0], img_size[1], 3)),

    # Prapemrosesan dan Augmentasi di dalam model
    Rescaling(1./255),
    data_augmentation_layers,

    # Blok Konvolusi 1
    Conv2D(32, (3,3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(2,2),

    # Blok Konvolusi 2
    Conv2D(64, (3,3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(2,2),

    # Blok Konvolusi 3
    Conv2D(128, (3,3), padding='same'),
    BatchNormalization(),
    Activation('relu'),
    MaxPooling2D(2,2),

    # Blok Klasifikasi
    Flatten(),
    Dense(128),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.5), # Regularisasi untuk mencegah overfitting
    Dense(1, activation='sigmoid') # Output biner (NORMAL atau PNEUMONIA)
])

# Kompilasi model
model.compile(
    optimizer=Adam(learning_rate=0.001), # Learning rate awal bisa lebih tinggi karena ada scheduler
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()


# ==============================================================================
# 6. STRATEGI 3: MELATIH MODEL DENGAN CALLBACKS CANGGIH
# ==============================================================================
# Definisikan callbacks untuk training yang lebih pintar
callbacks = [
    # Hentikan training jika tidak ada kemajuan dan kembalikan bobot terbaik
    EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True),
    # Kurangi learning rate secara otomatis saat performa stagnan
    ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=2, verbose=1, min_lr=1e-6)
]

# Mulai proses training
history = model.fit(
    train_generator,
    epochs=25, # Jumlah epoch maksimal, EarlyStopping akan berhenti jika perlu
    validation_data=val_generator,
    callbacks=callbacks,
    class_weight=class_weights # Terapkan bobot kelas di sini
)


# ==============================================================================
# 7. EVALUASI MODEL
# ==============================================================================
print("\n--- Mengevaluasi Performa Model ---")
# Evaluasi pada data test (model sudah otomatis menggunakan bobot terbaik berkat restore_best_weights=True)
test_loss, test_acc = model.evaluate(test_generator, verbose=1)
print(f"\nAkurasi pada Test Set: {test_acc*100:.2f}%")
print(f"Loss pada Test Set: {test_loss:.4f}")

# Plot grafik akurasi dan loss
plt.figure(figsize=(14, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.tight_layout()
plt.show()

# Laporan Klasifikasi yang detail
print("\n--- Laporan Klasifikasi ---")
y_true = np.concatenate([y for x, y in test_generator], axis=0).flatten()
y_pred_probs = model.predict(test_generator)
y_pred_labels = (y_pred_probs > 0.5).astype(int).flatten()

print(classification_report(y_true, y_pred_labels, target_names=class_names))
print("\n--- Confusion Matrix ---")
print(confusion_matrix(y_true, y_pred_labels))


# ==============================================================================
# 8. PREDIKSI PADA GAMBAR ACAK
# ==============================================================================
def predict_random_images(
    model,
    test_dir,
    class_names,
    img_size=(150, 150),
    num_images=10
):
    """Fungsi untuk memprediksi dan menampilkan hasil pada gambar acak."""
    normal_dir = os.path.join(test_dir, 'NORMAL')
    pneumonia_dir = os.path.join(test_dir, 'PNEUMONIA')

    all_image_paths = []
    # Ambil setengah gambar dari masing-masing kategori
    for category_dir in [normal_dir, pneumonia_dir]:
        if os.path.exists(category_dir):
            files = os.listdir(category_dir)[:num_images // 2]
            all_image_paths.extend([os.path.join(category_dir, f) for f in files])

    if not all_image_paths:
        print("Tidak dapat menemukan gambar di direktori test.")
        return

    random.shuffle(all_image_paths)

    plt.figure(figsize=(20, 10))
    for i, img_path in enumerate(all_image_paths):
        # Load gambar
        img = image.load_img(img_path, target_size=img_size)
        img_array = image.img_to_array(img)
        # Tambahkan dimensi batch
        img_array_expanded = np.expand_dims(img_array, axis=0)
        # TIDAK PERLU RESCALE DI SINI, karena model sudah punya layer Rescaling

        # Prediksi
        pred_prob = model.predict(img_array_expanded, verbose=0)[0][0]
        pred_label = class_names[1] if pred_prob > 0.5 else class_names[0]
        color = 'red' if pred_label == 'PNEUMONIA' else 'green'

        # Tampilkan
        plt.subplot(2, num_images // 2, i + 1)
        plt.imshow(img)
        true_label = "PNEUMONIA" if "PNEUMONIA" in img_path else "NORMAL"
        plt.title(f"Asli: {true_label}\nPrediksi: {pred_label}\nConfidence: {pred_prob:.2f}", color=color)
        plt.axis('off')

    plt.suptitle("Prediksi Model pada Gambar Acak dari Test Set", fontsize=16)
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()

# Panggil fungsi prediksi
predict_random_images(model, test_dir, class_names)

# Simpan model final
model.save('pneumonia_detection_optimized_model.h5')
print("\nModel final telah disimpan sebagai 'pneumonia_detection_optimized_model.h5'")
```
